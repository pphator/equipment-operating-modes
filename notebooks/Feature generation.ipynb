```python
import pandas as pd
import numpy as np
```

```python
data_path = '..\\data\\'
```

```python
X_train_val = np.load(data_path + 'X_train.npy')
y_train_val = np.load(data_path + 'y_train.npy')
X_test = np.load(data_path + 'X_test.npy')
y_test = np.load(data_path + 'y_test.npy')
```

Возьмём одну серию наблюдений для описания генерируемых предикторов

```python
sample = pd.DataFrame(X_train_val[12])
```

```python
sample
```

Посчитаем базовые статистики, которые станут первыми признаками

```python
sample.describe()
```

Сделаем разностное преобразование первого порядка

```python
sample_diff = sample.diff().dropna()
```

```python
sample_diff
```

Базовые статистики над разностным преобразованием сигналов также включим в список предикторов

Добавим вил тренда для сигнала с каждого датчика как ещё один признак

Для этого сначала посчитаем скользящее среднее для окна в 20 наблюдений

```python
rolling_mean = sample.rolling(window=20).mean().dropna()
```

```python
rolling_mean
```

Определим восходящий тренд: последнее наблюдение больше первого, и они отличаются более чем на 2 стандартных отклонения скользящей средней <br>
Определим нисходящий тренд: последнее наблюдение меньше первого, и они отличаются более чем на 2 стандартных отклонения скользящей средней <br>
Иначе фиксируем отсутствие тренда

```python
trend_series = rolling_mean.apply(lambda column: 'upward' if column.iloc[-1] - column.iloc[0] > 0 else 'downward' if np.abs(column.iloc[-1] - column.iloc[0]) > 2*column.std() else 'flat', axis=0)
```

```python
trend_series
```

Обучим One-hot encoder кодировать эти значения

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_encoder = OneHotEncoder()
```

```python
encoder_fit_array = trend_series.values[:, np.newaxis]
```

```python
encoder_fit_array
```

На выходе кодировщика получим матрицу

```python
print(oh_encoder.fit_transform(encoder_fit_array).toarray())
```

Рассчитаем признаки для каждой серии наблюдений

```python
features_train_val = []
for i in range(X_train_val.shape[0]):
    sample = pd.DataFrame(X_train_val[i])
    mean_values = np.mean(sample, axis=0)
    std_values = np.std(sample, axis=0)
    min_values = np.min(sample, axis=0)
    max_values = np.max(sample, axis=0)
    min_max_diff = max_values - min_values
    
    sample_diff = pd.DataFrame(sample).diff().dropna()
    sample_diff_mean = sample_diff.mean(axis=0).values
    sample_diff_std = sample_diff.std(axis=0).values
    sample_diff_max = sample_diff.max(axis=0).values
    sample_diff_min = sample_diff.min(axis=0).values
    sample_diff_max_min_range = sample_diff_max - sample_diff_min
    
    rolling_mean = sample.rolling(window=20).mean().dropna()
    trend_series = rolling_mean.apply(lambda column: 'upward' if column.iloc[-1] - column.iloc[0] > 0 else 'downward' if np.abs(column.iloc[-1] - column.iloc[0]) > 2*column.std() else 'flat', axis=0)
    trend_series_values = trend_series.values[:, np.newaxis]
    
    trend_features = oh_encoder.transform(trend_series_values).toarray()
    trend_features = trend_features.reshape(-1)
    
    sample_features = np.hstack((mean_values, std_values, min_values, max_values, min_max_diff, 
                                 sample_diff_mean, sample_diff_std, sample_diff_min, sample_diff_max,
                                sample_diff_max_min_range, trend_features))
    
    features_train_val.append(sample_features)
features_train_val = np.stack(features_train_val)
```

Проверим, что размерность совпадает

```python
assert features_train_val.shape[0] == X_train_val.shape[0]
```

Сохраним получившийся набор данных в файл

```python
np.save(data_path + 'features_train_val.npy', features_train_val)
```
