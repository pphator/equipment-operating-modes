Посчитаем среднюю корреляцию на обучающей истории наблюдений

```python
total_corr = np.zeros((X_train.shape[2], X_train.shape[2]))
for i in range(X_train.shape[0]):
    sample_df = pd.DataFrame(X_train[i])
    total_corr = total_corr + sample_df.corr()
total_corr = total_corr/X_train.shape[0]
```

```python
total_corr
```

```python
sns.heatmap(total_corr, vmin=-1)
```

Изучим возможность определения тренда в значениях в рамках каждой серии наблюдений

```python
for each_column in sample_df.columns:
    trend = 'no trend'
    sample_df[each_column].plot(color='g')
    rolling_mean = sample_df[each_column].rolling(window=20).mean().dropna()
    rolling_mean.plot(color='r')
    first_value = rolling_mean.iloc[0]
    last_value = rolling_mean.iloc[-1]
    if (last_value > first_value) and (np.abs(last_value - first_value) > 2*rolling_mean.std()):
        trend = 'upward'
    elif (last_value < first_value) and (np.abs(last_value - first_value) > 2*rolling_mean.std()):
        trend = 'downward'
    else:
        trend = 'flat'
    plt.title('Trend = {}'.format(trend))
    plt.show()
```
